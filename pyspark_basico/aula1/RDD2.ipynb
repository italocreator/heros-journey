```python
# Carregando dados
data = sc.textFile("regiao-norte.txt,regiao-sul.txt")
```

```python
# Separando Palavras
palavras = data.flatMap(lambda x: x.split(" "))
```

```python
for plv in palavras.collect():
    print(plv)
```

```python
#Funcao que tranforma em minuscula todas as palavras
def lowercase_word(x):
     minuscula = x.lower()
     return (minuscula)
lower_words = palavras.map(lambda x: lowercase_word(x))
```

```python
lower_words.collect()
```

```python
# utilizando stop words
stop_words = ['e','do','das','na','há','ser','são','para','por', 'as','em','no','nas','uma','que','como','com','','os','o','a','da','de','nas','é']
filter_words = lower_words.filter(lambda x: x not in stop_words)
```

```python
filter_words.collect()
```

```python
#Praparando o contador, (Palavra, 1)
freq_words = filter_words.map(lambda x: [x,1]) 
```

```python
#Frequencia de palavras
freq_words.collect()
```

```python
#ReduceByKey para somar o contador
from operator import add
total_words = freq_words.reduceByKey(add)
```

```python
total_words.collect()
```

```python
#Realizando um for interativo para colocar as palavras em ordem decescentes
for i in total_words.sortBy(keyfunc = lambda x:x[1],ascending=False).collect():print(i)
```
